## 4. Conceptos básicos

### Inteligencia Artificial

La **Inteligencia Artificial** se refiere al desarrollo de sistemas capaces de realizar tareas que normalmente requieren inteligencia humana, como reconocer patrones, tomar decisiones, interpretar información o generar respuestas.

### Machine Learning

El **Machine Learning** o aprendizaje automático es una rama de la inteligencia artificial que permite construir modelos a partir de datos. En lugar de programar todas las reglas de forma explícita, el modelo aprende patrones a partir de ejemplos.

### Tipos principales de aprendizaje automático

| Tipo de aprendizaje | Pregunta principal | Ejemplo en área química |
|---|---|---|
| Regresión | ¿Cuánto vale una variable numérica? | Predicción de punto de ebullición, viscosidad o conversión |
| Clasificación | ¿A qué categoría pertenece una muestra? | Agua potable/no potable, producto aceptable/no aceptable |
| Clustering | ¿Existen grupos naturales en los datos? | Agrupamiento de compuestos, muestras ambientales o condiciones de proceso |
| Detección de anomalías | ¿Este comportamiento es normal o inusual? | Detección de fallas en sensores o desviaciones de operación |
